# Choosing a classification algorithm
- "No Free Lunch": no single clf works best across all possible scenarios.
- At least try a handful of algorithms to select the best model, which depends on...
    - n_features
    - n_samples
    - noise
    - linearly separable or not

**5 main steps**
1. feature selection
2. choose performance metric
3. choose a clf and optimization algo
4. model evaluation
5. parameter tuning

# Perceptron via sklearn: miminize misclassification error

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
X = iris.data[:, [2,3]] # petal lenth, petal width, 150 samples
y = iris.target
```

```python
np.unique(y)
```

## train-test split

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

## Standardize features

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train) # use the same scaling param to scale the test set
```

```python
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test) # use the same scaling param to scale the test set
```

## Training
- One-vs.-Rest (OvR) method as default in sklearn

```python
from sklearn.linear_model import Perceptron
ppn = Perceptron(n_iter=40, eta0=0.1, random_state=0)
ppn.fit(X_train_std, y_train)
```

## Prediction

```python
y_pred = ppn.predict(X_test_std)
print('Misclassified samples: %d' % (y_test != y_pred).sum())
```

## Metric

```python
from sklearn.metrics import accuracy_score
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
```

## Visualize decision boundary

```python
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=.02):
    
    # setup marker generator and color map
    markers = ('s','x','o','^','v')
    colors = ('red','blue','lightgreen','gray','cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min()-1, X[:, 0].max()+1 # get min and max for each feature
    x2_min, x2_max = X[:, 1].min()-1, X[:, 1].max()+1
    
    # we want to just create a very fine scatter plot for each grid point
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), # np.arange(min,max,bin): similar to linspace
                          np.arange(x2_min, x2_max, resolution)) # create a mesh grid (shape: len(xx1), len(xx2))
    
    # prediction for each grid point
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    
    plt.contourf(xx1, xx2, Z, alpha=.2, cmap=cmap)
    plt.xlim(xx1.min(),xx1.max())
    plt.ylim(xx2.min(),xx2.max())
    
    # plot class samples (actual data)
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x = X[y==cl, 0], y = X[y==cl, 1], 
                    alpha=0.8, c=cmap(idx),
                    marker=markers[idx], label=cl)
        
    """MODIFICATION: highlight test samples """
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:, 0], X_test[:, 1], 
                    c='', alpha=1.0, linewidths=1, marker='o', s=55, label='test set')
        # c='' makes the marker color empty
        # s: size
```

```python
# combining train and test for plotting
X_combined_std = np.vstack((X_train_std, X_test_std)) 
# np.vstack: Take a sequence of arrays and stack them vertically to make a single array.
# pretty convenient method!

y_combined = np.hstack((y_train, y_test))
```

```python
X_combined_std.shape, y_combined.shape
```

```python
plot_decision_regions(X=X_combined_std, y=y_combined, 
                      classifier=ppn, test_idx=range(y_train.shape[0],y_combined.shape[0]))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```

- Perceptron can't do a great job because the data is not linearly separable.
- In fact, perceptron never converges on datasets that aren't perfectly linearly separable.

# Logistic regression: modeling class prob 

## Logistic regression intuition
- linear model for binary classification that can be extended to multiclass 
- odds ratio = p / (1 - p) where p is the prob of positive event
- logit(p) = log ( p / (1 - p) )
- logit(p(y=1|x) = wT x (linear combination)
- logistic: inverse of logit
    - phi_z = 1 / (1 + e^(-z)): z = linear combination (wTx)
- we're interested in predicting phi_z, the probability!

```python
# visualize sigmoid
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

z = np.arange(-7, 7, 0.1)
phi_z = sigmoid(z)
```

```python
plt.plot(z, phi_z)
plt.axvline(0.0, color='k') # add a vertical line
plt.axhspan(0.0, 1.0, facecolor='1.0', alpha=1.0, ls='dotted') # add horizontal span (rectangle)
# matplotlib facecolor float: grayscale (1.0 white, 0.0 black)
# ls = linestyle
plt.axhline(y=0.5, ls='dotted', color='k')
plt.yticks([0.0, 0.5, 1.0])
plt.ylim([-0.1, 1.1])
plt.xlabel('z')
plt.ylabel('$\phi (z)$')
plt.show()
```

- Compared to Adaline, here the activation function is sigmoid function instead of identity function.
- Quantizer (unit step function) still exists. If phi_z > 0.5, then class 1, else 0.

## learning the weights of cost function
- In Adaline, we used SSE.
- Here, we calculate likelihood function. We want to maximize this.
- likelihood (L(w)) = product of p(y given x and weight)
    - Which is, product of (phi_z power to y_i) and (1-phi_z) power to (1-y)
- if we take log, logL = sum [y * log(phi_z) + (1-y) * log(1-phi_z)
- to use gradient descent, we can **rewrite** the logL as a cost function: **just add (-)**
- for a single sample:
    - cost = -y log(phi_z) - (1-y)log(1-phi_z)
    - if y = 1, cost = -log(phi_z)
    - if y = 0, cost = -log(1-phi_z)        

```python
Jw_y1 = -np.log(phi_z)
Jw_y0 = -np.log(1-phi_z)
```

```python
plt.plot(phi_z,Jw_y1,c='k',label='J(w) if y=1')
plt.plot(phi_z,Jw_y0,':',c='k',label='J(w) if y=0')
plt.xlabel('Probability, $\phi (z)$')
plt.ylabel('Cost, J(w)')
plt.legend(loc='upper center')
plt.show()
```

Here, phi_z is the probability of getting the class right (y=1). The cost reduces as our probility of predicting class as 1 if the class is actually 1. However, if the actual class is 0, if it becomes more certain (high phi_z) that the predicted class is 1, the cost increases.

## Training
- we now have a new cost function: **-sum(y log(phi_z) + (1-y)log(1-phi_z))**
- but we can simply use sklearn's logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=1000.0, random_state=0)
lr.fit(X_train_std, y_train)
```

```python
plot_decision_regions(X_combined_std, y_combined, classifier=lr, test_idx=range(y_train.shape[0],y_combined.shape[0]))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```

We can also predict the class-membership probaiblity of the samples via `predict_proba` method.

```python
lr.predict_proba(X_test_std[:2,:])
```

Weight update in logistic regression via gradient descent is equal to that of Adaline.
- We can prove this by using partial derivatives of log likelihood wrt the weight and the partial derivative of the sigmoid function wrt z.

## regularization: tackling overfitting
Regularization: finding a good bias-variance tradeoff
- good for handling collinearity
- filter out noise from data
- prevent overfitting

**Basic idea**: introduce additional information(**bias**) to penalize extreme parameter weights
- most common one, **L2 regularization** (aka L2 shrinkage or weight decay)
    - lambda/2 * sum(weight^2)
- for regularization, **feature scaling** is very important.

logistic regression parameter **C** = 1/lambda -> large C, weaker regularization

```python
lr.coef_[1] #(n_classes, n_features), Coefficient of the features in the decision function
```

```python
weights, params = [], []
for c in np.arange(-5, 5):
    lr = LogisticRegression(C=10**c, random_state=0)
    lr.fit(X_train_std, y_train)
    
    #(n_classes, n_features), Coefficient of the features in the decision function
    weights.append(lr.coef_[1]) # here, we're getting the weights for the second class
    params.append(10**c)

weights = np.array(weights)
```

```python
plt.plot(params, weights[:, 0], label='petal length')
plt.plot(params, weights[:, 1], ls='--', label='petal width')
plt.ylabel('weight coefficient')
plt.xlabel('C')
plt.legend(loc='upper left')
plt.xscale('log')
plt.show()
```

## self-study: how about model performance?

```python
lr.score(X_test_std, y_test)
```

```python
for c in np.arange(-5, 5):
    lr = LogisticRegression(C=10**c, random_state=0)
    lr.fit(X_train_std, y_train)
    accuracy_score = lr.score(X_test_std, y_test)
    
    plot_decision_regions(X_combined_std, y_combined, classifier=lr, test_idx=range(y_train.shape[0],y_combined.shape[0]))
    plt.xlabel('petal length [standardized]')
    plt.ylabel('petal width [standardized]')
    plt.title('C = ' + str(10**c))
    plt.legend(loc='upper left')
    plt.text(1.5, -2, 'Score: %.2f' %accuracy_score, fontsize=15)
    plt.show()
```

# SVM: maximizing marging classification
- margin: distance btw the separating hyperplane (decision boundary)
- support vectors: training samples that are closest to the hyperplane
- decision boundary: wTx = 0
    - negative hyperplane: wTx = -1, positive: wTx = 1
    
## Intuition
- Largest margin error -> good for generalization
- margin can be defined as distance between negative and positive hyperplane
    - positive - negative = 2, then we normalize this by ||w||
    - then the left hand side becomes the distance and the right side becomes 2 / ||w||    
- objective function
    - goal: maximization of the margin (2 / ||w||)
    - with constraints that the samples are classified correctly
        - if y = 1, wTx >= 1, if y = -1, wTx <= -1
- practically, it means we minimize the inverse, which is **||w||/2**

## Dealing with the nonlinearly separable case using <span class="mark">slack</span> variables
- linear constraints can be **relaxed** for nonlinearly separable data to **allow opimization convergence** -> **solf-margin**, with under **appropriate cost penalty**
- basically we are adding or subtracting slack variable to the constraint (1 or -1) to reduce the margin
    - if y = 1, wTx >= 1-xi, if y = -1, wTx <= -1+xi 
- objective to minimize becomesn:
    - 2/||w|| + C * sum(xi) <span class="mark">WHY?</span>
- C: control variable to penalize misclassification
    - **similar to regularlization**: larger C: larger error penalty
    - we can control the margin with C and tun the bias-variance trade-off

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear', C=1.0, random_state=0)
svm.fit(X_train_std, y_train)
```

```python
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(y_train.shape[0],y_combined.shape[0]))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```

```python
for c in np.arange(-3, 3):
    svm = SVC(kernel='linear', C=10**c, random_state=0)
    svm.fit(X_train_std, y_train)
    
    accuracy_score = svm.score(X_test_std, y_test)
    
    plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(y_train.shape[0],y_combined.shape[0]))
    plt.xlabel('petal length [standardized]')
    plt.ylabel('petal width [standardized]')
    plt.title('C = ' + str(10**c))
    plt.legend(loc='upper left')
    plt.text(1.5, -2, 'Score: %.2f' %accuracy_score, fontsize=15)
    plt.show()
```

# Kernel SVM: solving nonlinear problems
- why popular among ML practitioners: easily kernelized to solve nonlinear problems

**Creating a sample dataset using XOR gate**
- XOR: true if inputs are different
- use np.logical_xor

```python
np.logical_xor(True,False)
```

```python
np.random.seed(0)
X_xor = np.random.randn(200,2) # "standard normal" distribution rand vars
y_xor = np.logical_xor(X_xor[:, 0]>0, X_xor[:, 1]>0) # True for second and fourth quadrant (when the signs are different)
y_xor = np.where(y_xor, 1, -1) # convert boolean to int

plt.scatter(X_xor[y_xor==1, 0], X_xor[y_xor==1, 1], c='b', marker='x', label='1')
plt.scatter(X_xor[y_xor==-1, 0], X_xor[y_xor==-1, 1], c='r', marker='s', label='-1')
plt.ylim(-3.0)
plt.legend()
plt.show()
```

This is not linearly separable. We need kernel tricks.
- create nonlinear combinations of the original features to project them onto a higher dimensional space via mapping function phi, where it becomes linearly separable.

## kernel trick to find separating hyperplanes in higher dimensional place
Kernel approach
1. transform training data onto a higher dimensional feature space via a mapping function (phi)
2. train a linear SVM in this new feature space
3. transform the test dataset (by using the same mapping function) for classification

Potential problem: construction of new feature can be **computationally very expensive** (especially with high dimensionality

**This is where the kernel trick comes into play.**
- our goal: replace the dot product xi * xj by phi(xi) * phi (xj)
- to save this expensive step of calculating this dot product between two points explicitly, we define a kernel function
    - k(xi, xj) = phi(xi) * phi(xj)

**Radial Basis Function kernal (RBF) or aka Gaussian kernel**
- one of the most popular ones
- k(xi, xj) = exp (-gamma * ||xi - xj||)
    - gamma = 1/ (2 * sigma^2)
- can be interpreted as a **similarity function** betwee the sample pair
    - more similar -> higher value (when identical, it's 1)
    - more dissimillar -> lower, (closer to zero)

```python
for kernel_ in ['linear', 'rbf']:    
    svm = SVC(kernel=kernel_, random_state=0, gamma=0.10, C=10.0)
    svm.fit(X_xor, y_xor)
    plot_decision_regions(X_xor, y_xor, classifier=svm)
    plt.legend(loc='upper left')
    plt.title(kernel_)
    plt.show()
```

gamma: inverse of variance
- larger gamma -> smaller variance -> more penalty on dissimilarity
- **cut-off** parameter for the Gaussian sphere
    - smaller gamma -> increase the influence (or reach) of the training samples -> softer decision boundray
    
Effect of gamma
- if gamma is too small (large variance), most of the transformed variables will be more or less close to 1.
    - it's almost like a uniform distribution.
- if gamma is too large (small variance), most of the transformed varialbes will be zero except for the ones with high similarity.
    - it's almost like a delta function. -> can be overfitting

```python
for c in np.arange(-3,3):
    gamma = 10**c
    X_tf = np.exp((X_xor[:,0]-X_xor[:,1])**2*-gamma)
    plt.plot(X_tf[y_xor==1], 'o', c='b', label='1')
    plt.plot(X_tf[y_xor==-1], '^', c='r', label='-1')
    plt.legend(loc='lower left')
    plt.title('$\gamma$: '+str(10**c))
    plt.ylim([0,1])
    plt.show()
```

```python
for c in np.arange(-3,3):
    svm = SVC(kernel='rbf', random_state=0, gamma=10**c, C=10.0)
    svm.fit(X_xor, y_xor)
    plot_decision_regions(X_xor, y_xor, classifier=svm)
    plt.legend(loc='upper left')
    plt.title('$\gamma$: '+str(10**c))
    plt.show()
```

Applying to Iris data

```python
svm = SVC(kernel='rbf', random_state=0, gamma=0.2, C=1.0)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, svm, range(105,150))
plt.legend(loc='upper left')
plt.show()
```

```python
svm = SVC(kernel='rbf', random_state=0, gamma=100, C=1.0)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, svm, range(105,150))
plt.legend(loc='upper left')
plt.show()
```

If gamma is too big, decision boundary becomes too specific which can cause overfitting.

# Decision tree
- Great for interpretability
- start at the tree root and split the data on the feature that results **the largest information gain (IG)**
- repeat this process until the leaves (children) are pure (all same class)
    - This can produce very deep (overfitted) tree. So we need **pruning** (= set a limit for the max depth)

## Objective function: maximize information gain

### IG at each split for a feature is defined as:
[parent's impurity measure (I)] - sum [product of the ratio for each class and corresponding child's impurity measure)]
- ex. I(parent) - sum( N_class1 / N_total X I(split_class1) + N_class2 / N_total X I(split_class2)

### Impurity measure: Gini impurity (Ig), entropy (Ih), classification error (Ie)
1. Ih: - sum [p log2(p)]
    - if everything belongs to a same class (p=1): Ih = 0
    - if everything is uniformly split (ex. p=0.5): Ih = 1
2. Ig: 1 - sum(p^2)
    - a criterion to minimize the prob of misclassification
    - maximum when everything is perfectly mixed (uniform)
    - In practice, **Ig and Ih usually give similar results (not worth spending time tweaking this)**
3. Ie: 1 - max(p)
    - userful for pruning but not good for growing a tree (**less sensitive to changes in the class prob**)

## impurity measure comparison

```python
def gini(p):
    return 1 - (p**2 + (1-p)**2)
def entropy(p):
    return -p*np.log2(p) - (1-p)*np.log2(1-p)
def error(p):
    return 1-np.max([p, 1-p])
```

```python
x = np.arange(0,1,0.01)

ent = [entropy(p) if p !=0 else None for p in x] # if comes before for
sc_ent = [e*.5 if e else None for e in ent] # scaled entropy (*1/2)

err = [error(i) for i in x]
```

```python
fig = plt.figure()
ax = plt.subplot(111)
for i, lab, ls, c in zip([ent, sc_ent, gini(x), err], 
                         ['Entropy', 'Entropy (scaled)', 'Gini Impurity', 'Misclassification Error'], 
                         ['-', '-', '--', '-.'], 
                         ['black', 'lightgray', 'red', 'green', 'cyan']):
    line = ax.plot(x, i, label=lab, linestyle=ls, lw=2, color=c)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3, fancybox=True, shadow=False) #ncols: number of columns
ax.axhline(y=0.5, linewidth=1, color='k', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='k', linestyle='--')
plt.ylim([0, 1.1])
plt.xlabel('p(i=1)')
plt.ylabel('Impurity Index')
plt.show()
```

## Random forest: combining weak to strong learners
- better generalization, less susceptible to overfitting

Steps
1. randomly pick *n* samples (with replacement) from training set
2. grow a decision tree. For each node:
    1. randomly select *d* features without replacement
    2. split (grow a tree) only using these features
3. repeat 1 and 2, *k* times
4. aggregate the prediction by each tree to assign class labels based on **majority vote**.

Parameters
- typically don't need pruning much since ensemble model is robust to individual trees
- *k* matters (larger -> better performance)
- *n* and *d* can be optimized (less common)
    - **larger *n* **: less randomness -> more likely to overfit (default *n*: training size)
    
sklearn randomforest provides **n_jobs** parameters where you can **parallelize** the process for better performance. **n_jobs** corresponds to the number of cores.

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(criterion='entropy', 
                                n_estimators=10, # number of trees
                                random_state=1, 
                                n_jobs=2) # parallelize the process (number of cores)
forest.fit(X_train,y_train)
```

```python
plot_decision_regions(X_combined, y_combined, 
                      classifier=forest, test_idx=range(y_train.shape[0],y_combined.shape[0]))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```

# KNN: lazy learner
**Lazy**
- Doesn't learn a discriminative function
- Instead, memorize training data

Parametric learners
- estimate params from the training set to learn a function which can classify new data points
- training data not required once function is learned
- examples: perceptron, logistic regression, linear SVM

Nonparametric learners
- can't be characterized by a fixed set of parameters
- n_param grows with traning data
- examples: decision tree, random forest, kernel SVM, KNN

KNN: **instance-based learning**
- memorize the training dataset
- *lazy learning is a specific case of instance-based learning*: **no cost during the learning process**

How KNN works
1. Choose the number of k and a distance metric
2. Find the k nearest neighbors of the sample that we want to classify
3. Assign the class label by majority vote.

**Memory-based model**
- Pros
    - classifier adapts as soon as new data is collected
- Cons
    - computational complexity for classifying new samples grows linearly with training dataset size (in the worst case)
    - can't discard training samples (storage issue)

<span class="mark">**Curse of dimensionality**</span>
- KNN is especially susceptible to this.
- Feature space becomes increasingly sparse for an increasing number of dimensions of a fixed-size training dataset
- Logistic regression can use regularization to prevent overfitting but KNN doesn't have such
- So for KNN, you need to feature selection and dimensionality reduction.

## Implement

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski') # p=2, metric='minkowski' => Euclidean
knn.fit(X_train_std,y_train)
```

```python
plot_decision_regions(X_combined_std, y_combined, 
                      classifier=knn, test_idx=range(y_train.shape[0],y_combined.shape[0]))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```

Param tuning
- right choice of k -> under-/over-fitting
- good metric
    - Euclidean: need to be standardized
