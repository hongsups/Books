# PCA
Intro
- Helps identify patterns in data based on feature correlation
- Finds the directions of max variance in high-dimensional data and projects it onto a new subspace
- The orthogonal axes (principal components) can be interpreted as the directions of max variance **given the constraint that the new feature axes are orthogonal to each other**

Idea
- d x k matrix transformation (d: original, k: transformed, typically k << d)

Warning: highly sensitive to scaling

Steps
1. Standardize the dataset (d-dimension)
2. Construct the covariance matrix
3. Decompose the covariance matrix into its eigenvectors and eigenvalues
4. Select k eigenvectors that correspond to the k largest eigenvalues (k: new feature subspace dim (k<=d))
5. Construct a projection matrix W from the "top" k eigenvectors.
6. Transform the input (d-dimension) using the projection matrix W to obtain the new feature subspace (k-dimension)

```python
import pandas as pd
df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',header=None)
```

```python
# train-test split and standardize
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler

X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.fit_transform(X_test)
```

- Covariance matrix show how features are correlated (increase together, etc.).
- Note that after standardizing, average of each feature vector is zero.
- Eigenvector direction: direction of the max variance, value: magnitude
- Eigenvector calculation can be done by using numpy `linalg.eig`

```python
import numpy as np
cov_mat = np.cov(X_train_std.T) # why transpose? this way, you can get d x d matrix (other wise, you will get n_rows x n_rows matrix)
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)
print('Eigenvalues \n%s' % eigen_vals) # \n: new line
```

Note: `linalg.eig` can return complex numbers. `linalg.eigh` can return more stable results (always return real eigenvalues). See [Hermitian matrix](https://en.wikipedia.org/wiki/Hermitian_matrix).

## Variance explained
Var_explained = the fraction of an eigenvalue and the total sum of the eigenvalues

```python
tot = sum(eigen_vals)
var_exp = [(i/tot) for i in sorted(eigen_vals, reverse=True)] # we want to sort from the largest
cum_var_exp = np.cumsum(var_exp)
```

## PCA in sklearn
Example: PCA to transform variables, use logistic regression to classify labels, and visualize

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=.02):
    
    # setup marker generator and color map
    markers = ('s','x','o','^','v')
    colors = ('red','blue','lightgreen','gray','cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min()-1, X[:, 0].max()+1 # get min and max for each feature
    x2_min, x2_max = X[:, 1].min()-1, X[:, 1].max()+1
    
    # we want to just create a very fine scatter plot for each grid point
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), # np.arange(min,max,bin): similar to linspace
                          np.arange(x2_min, x2_max, resolution)) # create a mesh grid (shape: len(xx1), len(xx2))
    
    # prediction for each grid point
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    
    plt.contourf(xx1, xx2, Z, alpha=.2, cmap=cmap)
    plt.xlim(xx1.min(),xx1.max())
    plt.ylim(xx2.min(),xx2.max())
    
    # plot class samples (actual data)
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x = X[y==cl, 0], y = X[y==cl, 1], 
                    alpha=0.8, c=cmap(idx),
                    marker=markers[idx], label=cl)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
lr = LogisticRegression()

X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std) # again, important thing: you fit only for the training set and transform
lr.fit(X_train_pca, y_train)

plot_decision_regions(X_train_pca, y_train, lr)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(loc='lower left')
plt.show()
```

It's a mirror image to the previous plot. This is due to eigenvectors' sign.

```python
plot_decision_regions(X_test_pca, y_test, lr)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Test set')
plt.legend(loc='lower left')
plt.show()
```

To <span class="mark">get explained variance ratio</span>, we can set `n_components=None` and use all components.

```python
pca = PCA(n_components=None)
X_train_pca = pca.fit_transform(X_train_std)
pca.explained_variance_ratio_
```

# Linear discriminant analysis (LDA) for supervised data compression
Idea
- Find the feature subspace that **optimizes class separability**

LDA vs. PCA
- Both are linear transformation techniques but LDA is supervised.
- Then, PCA is better than LDA? Not quite. Some studies found that PCA is better for image recognition task with small number of samples.

**Assumptions**: all though but LDA can still work with slight violation
- Normally distributed data
- Identical covariance matrix of classes
- Statistically independent features

Steps
1. Standardize the d-dim input 
2. For each class compute the d-dim mean vector
3. Construct the between-class scatter matrix S_B and the within-class scatter matrix S_W
4. Compute the eigenvectors and corresponding eigenvalues of the matrix S-1_W * S_B
5. Choose the k eigenvectors that correspond to the k largest eigenvalues to construct d x k transformation matrix W; the eigenvectors are the columns of the this matrix
6. Project the sample onto the new feature subspace using W (transformation)

## Computing the scatter matrices
First, compute the mean vectors

```python
np.set_printoptions(precision=4)
mean_vecs=[]

for label in range(1,4): # loop over each class
    mean_vecs.append(np.mean(
        X_train_std[y_train==label], axis=0))
    print('MV %s: %s\n' %(label, mean_vecs[label-1]))
```

Compute the within-class scatter matrix = sum [individual scatter matrix of each class] for all class labels

```python
d = 13 # n_features
S_W = np.zeros((d,d)) # preallocate?

for label, mv in zip(range(1,4), mean_vecs):
    class_scatter = np.zeros((d, d))
    
    for row in X_train[y_train == label]: # for each row that correpond to a class
        row, mv = row.reshape(d, 1), mv.reshape(d, 1)
        class_scatter += (row-mv).dot((row-mv).T)
        
    S_W += class_scatter
print('Within-class scatter matrix: %sx%s' % (S_W.shape[0], S_W.shape[1]))        
```

Here, we made an implicit assumption: class density is uniform, which may not be true.

```python
print('Class label distribution: %s' % np.bincount(y_train)[1:]) # another convenient function
```

Now that there's imbalance, we need to scale the individual scatter matrix meaning we need to divide the dot product by the number of samples per class, which becomes the covariance matrix.

```python
d = 13  
S_W = np.zeros((d, d))
for label, mv in zip(range(1, 4), mean_vecs):
    class_scatter = np.cov(X_train_std[y_train == label].T)
    S_W += class_scatter
print('Scaled within-class scatter matrix: %sx%s' % (S_W.shape[0],S_W.shape[1]))
```

Now we compute the between-class scatter matrix.

```python
mean_overall = np.mean(X_train_std, axis=0)
d = 13  # number of features
S_B = np.zeros((d, d))
for i, mean_vec in enumerate(mean_vecs):
    n = X_train[y_train == i + 1, :].shape[0]
    
    mean_vec = mean_vec.reshape(d, 1)  # make column vector
    mean_overall = mean_overall.reshape(d, 1)  # make column vector
    S_B += n * (mean_vec - mean_overall).dot((mean_vec - mean_overall).T)

print('Between-class scatter matrix: %sx%s' % (S_B.shape[0], S_B.shape[1]))
```

## Select linear discriminants for the new feature space
The remaing steps are similar to those of the PCA but instead of the eigendecomposition on the covariance matrix, we solve the generalized eigenvalue problem of the matrix S_W(inv)S_B

```python
eigen_vals, eigen_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B))
```

```python
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:,i]) for i in range(len(eigen_vals))] 
eigen_pairs = sorted(eigen_pairs, key=lambda k: k[0], reverse=True) # sorted can use key and key can be retrieved through lambda

print('Eigenvalues in decreasing order: \n')
for eigen_val in eigen_pairs:
    print(eigen_val[0])
```

The number of linear discrminants are at most c-1 (c=n_class); that's why we have 2 non-zero eigenvalues.

```python
tot = sum(eigen_vals.real) # they are complex nubmers 
discr = [(i/tot) for i in sorted(eigen_vals.real, reverse=True)]
cum_discr = np.cumsum(discr)

plt.bar(range(1,14), discr, alpha=0.5, align='center', label='individual "discriminability"')
plt.step(range(1,14), cum_discr, where='mid', label='cumulative "discriminability"') 

plt.xlabel('Linear discriminants')
plt.ylabel('"discriminability" ratio')
plt.legend(loc='best')
plt.show()
```

Stack the two most discriminative eigenvector columns to create the transformation matrix W

```python
w = np.hstack((eigen_pairs[0][1][:, np.newaxis].real, # by adding np.newaxis (alias for None), we change a row vector to a column one
               eigen_pairs[1][1][:, np.newaxis].real))
print('Matrix W:\n',w)
```

## Project samples onto the new feature space

```python
X_train_lda = X_train_std.dot(w)
```

```python
colors = ['r','b','g']
markers = ['s','x','o']
for l, c, m in zip(np.unique(y_train), colors, markers):
    plt.scatter(X_train_lda[y_train==l, 0],
                X_train_lda[y_train==l, 1], label=l, c=c, marker=m)
plt.xlabel('LD 1')
plt.ylabel('LD 2')
plt.legend(loc='lower right')
plt.show()
```

# Kernel PCA: nonlinear mapping
For nonlinear data, PCA or LDA might not be an ideal choice.

## Kernel trick
**Idea of Kernel PCA**
- Using a nonlinear mapping, transform data to nonlinear higher dimension, then use standard PCA to lower the dimension back.
- As it sounds, this is **computationally very expensive**. Thus, we use **kernel trick.**

**Kernel trick**
- we can compute the similarity between two high-dimension feature vectors in the original feature space.
- **Avoid calculating the pairwise dot product of the samples under the nonlinear function explicitly by using a kernel function K**
- No need to calculate the eigenvectors explicitly.

**Kernel PCA outcome** 
- **Already projected onto the respective components**
- Not like PCA where we get the transformation matrix W and we needed to transform the input

**Common kernel tricks**
- polynomial: (product + threshold)^power
- tanh: tanh(eta x product + threshold)
- RBF (Gaussian)

**Implementing RBF; Steps**
1. Compute the kernel matrix k
    - Do this for each pair of sample (ex. if we have 100 samples, dim K will be 100 x 100)
2. Center the k matrix
    - Need to make sure the NEW feature space is also centered at zero (like we did for the standardized input)
3. Collect the top k eigenvectors of the centered kernel matrix based on the corresponding eigenvalues (descending order)
    - In contrast to PCA, the eigenvectors are **NOT the principal component axes** but <span class="mark">the samples projected onto those axes.</span>

## Implementing kernel PCA

```python
from scipy.spatial.distance import pdist, squareform
from scipy import exp
from scipy.linalg import eigh
import numpy as np
```

```python
def rbf_kernel_pca(X, gamma, n_components):
    """
    RBF kernel PCA implmentation.
    
    Parameters
    ----------
    X: {Numpy ndarray}, shape = [n_samples, n_features]
    
    gamma: float
        Tuning parameter of the RBF kernel
    
    n_components: int
        Number of principal components to return
    
    Returns
    -------
    X_pc: {Numpy ndarray}, shape = [n_samples, k_features]
        Projected dataset
    """
    
    # Calculate pairwise squared Euclidean distances in the MxN dimentional dataset.
    sq_dists = pdist(X, 'sqeuclidean') # (n_samples x n_features, 1)
    
    # Convert pairwise distances into a square matrix
    mat_sq_dists = squareform(sq_dists) # (n_samples x n_samples)
    
    # Compute the symmetric kernel matrix
    K = exp(-gamma * mat_sq_dists) # (n_samples x n_samples)
    
    # Center the kernel matrix.
    N = K.shape[0]
    one_n = np.ones((N,N))/N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)
    
    # Obtaining eigenpairs from the centered kernel matrix
    # numpy.eigh returns them in sorted order
    eigvals, eigvecs = eigh(K) # eigvecs: (n_sampels x n_samples)
    
    # Collect the top k eigenvectors (projected samples)
    X_pc = np.column_stack((eigvecs[:, -i] for i in range(1, n_components + 1))) # (n_samples x n_components)
    # note there here the range starts from 1 not 0
    # and counting backwards

    return X_pc
```

One downside: gamma paremter should be tuned.

np.column_stack: consider it as a hstack of column vectors (but even if the input vector is row, it converts to column and stack horizontally)

## Projecting new data points
- We will need to project the data that is not the training set.
- Kernel PCA is memory-based (not like standard PCA). So we have to **reuse the original training set each time to project new samples.**

```python
def rbf_kernel_pca(X, gamma, n_components):
    """
    RBF kernel PCA implmentation.
    
    Parameters
    ----------
    X: {Numpy ndarray}, shape = [n_samples, n_features]
    
    gamma: float
        Tuning parameter of the RBF kernel
    
    n_components: int
        Number of principal components to return
    
    Returns
    -------
    X_pc: {Numpy ndarray}, shape = [n_samples, k_features]
        Projected dataset
        
    lambdas: list
        Eigenvalues
    """
    
    # Calculate pairwise squared Euclidean distances in the MxN dimentional dataset.
    sq_dists = pdist(X, 'sqeuclidean') # (n_samples x n_features, 1)
    
    # Convert pairwise distances into a square matrix
    mat_sq_dists = squareform(sq_dists) # (n_samples x n_samples)
    
    # Compute the symmetric kernel matrix
    K = exp(-gamma * mat_sq_dists) # (n_samples x n_samples)
    
    # Center the kernel matrix.
    N = K.shape[0]
    one_n = np.ones((N,N))/N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)
    
    # Obtaining eigenpairs from the centered kernel matrix
    # numpy.eigh returns them in sorted order
    eigvals, eigvecs = eigh(K) # eigvecs: (n_sampels x n_samples)
    
    # Collect the top k eigenvectors (projected samples)
    alphas = np.column_stack((eigvecs[:, -i] for i in range(1, n_components + 1))) # (n_samples x n_components)

    """From here, it's different"""
    lambdas = [eigvals[-i] for i in range(1,n_components+1)]

    return alphas, lambdas
```

```python
X, y = make_moons(n_samples=100, random_state=123)
alphas, lambdas = rbf_kernel_pca(X, gamma=15, n_components=1)
```

Let's assume that 26th point from the dataset is the new data point. Let's project it onto the new subspace.

```python
x_new = X[25]
x_new
```

```python
x_proj = alphas[25]
x_proj
```

```python
def project_x(x_new, X, gamma, alphas, lambdas):
    pair_dist = np.array([np.sum((x_new-row)**2) for row in X])
    k = np.exp(-gamma * pair_dist)
    return k.dot(alphas/lambdas)
```

```python
x_reproj = project_x(x_new, X, 15, alphas, lambdas)
x_reproj
```

```python
plt.scatter(alphas[y==0,0], np.zeros((50)), color='red', marker='^', alpha=.5)
plt.scatter(alphas[y==1,0], np.zeros((50)), color='blue', marker='^', alpha=.5)

plt.scatter(x_proj, 0, color='black', label='original projection of point X[25]', marker='^', s=100)
plt.scatter(x_reproj, 0, color='green', label='remapped point X[25]', marker='x', s=500)

plt.legend(scatterpoints=1)
plt.show()
```

# Manifold learning

Sklearn provides advanced techniques for nonlinear dimensionality reduction called [**manifold learning**](http://scikit-learn.org/stable/modules/manifold.html).

**Intro** (from sklearn)
- Manifold Learning can be thought of as an attempt to **generalize linear frameworks like PCA to be sensitive to non-linear structure** in data. Though supervised variants exist, the typical manifold learning problem is **unsupervised**: it learns the high-dimensional structure of the data from the data itself, without the use of predetermined classifications.
- Make sure the same scale is used over all features. Because manifold learning methods are based on a nearest-neighbor search, the algorithm may perform poorly otherwise.

Here, I played with two examples, isomap and MDS.

```python
from sklearn import manifold
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X_std = sc.fit_transform(X)
```

```python
X_isomap = manifold.Isomap(n_neighbors=10, n_components=2).fit_transform(X)

plt.scatter(X_isomap[y==0,0], X_isomap[y==0,1], color='red', marker='^', alpha=.5)
plt.scatter(X_isomap[y==1,0], X_isomap[y==1,1], color='blue', marker='^', alpha=.5)
plt.title("Isomap")
plt.axis('tight')
plt.show()
```

```python
X_mds = manifold.MDS(n_components=2, max_iter=300, n_init=1).fit_transform(X)

plt.scatter(X_mds[y==0,0], X_mds[y==0,1], color='red', marker='^', alpha=.5)
plt.scatter(X_mds[y==1,0], X_mds[y==1,1], color='blue', marker='^', alpha=.5)
plt.title("MDS")
plt.axis('tight')
plt.show()
```
